```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
print(cv2.data.haarcascades) 
```

```python
img = cv2.imread(r'D:\OneDrive\Pictures\Screenshots\open cv\group.jpg')
img_2 = img.copy()
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

faces_rectangle = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3)
faces_rectangle_2 = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=15)

for(x,y,w,h) in faces_rectangle:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), thickness=5)

for(x,y,w,h) in faces_rectangle_2:
    cv2.rectangle(img_2, (x,y), (x+w,y+h), (0,255,0), thickness=5)

print(f'{4*'\t'}{len(faces_rectangle)} {7*'\t'}{len(faces_rectangle_2)}')
plt.figure(figsize=[10,5])
plt.subplot(121); plt.imshow(img[..., ::-1]); plt.title('minNeighbors=3')
plt.subplot(122); plt.imshow(img_2[..., ::-1]); plt.title('minNeighbors=15')
plt.tight_layout()
plt.show()
```
